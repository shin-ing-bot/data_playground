```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# 파일 업로드
uploaded = files.upload()
filename = list(uploaded.keys())[0]
print(f"업로드된 파일: {filename}")


df = pd.read_csv(filename)
print("\n영화 데이터 미리보기")
print(df.head(), "\n")


# 장르별 평점 평균
genre_rating = df.groupby("장르")["평점"].mean().reset_index().sort_values("평점", ascending=False)
print("장르별 평점 평균")
print(genre_rating.to_string(index=False), "\n")

# 장르별 평균 관객수
genre_audience = df.groupby("장르")["관객수"].mean().reset_index().sort_values("관객수", ascending=False)
print("장르별 평균 관객수")
print(genre_audience.to_string(index=False), "\n")

# 영화별 관객수 (내림차순 정렬)
movie_audience = df[["제목", "관객수"]].sort_values("관객수", ascending=False)
print(" 영화별 관객수")
print(movie_audience.to_string(index=False), "\n")

# 영화별 평점 (내림차순 정렬)
movie_rating = df[["제목", "평점"]].sort_values("평점", ascending=False)
print("영화별 평점")
print(movie_rating.to_string(index=False), "\n")

!apt-get -qq install fonts-nanum
import matplotlib.font_manager as fm
font_path = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"
fm.fontManager.addfont(font_path)
plt.rc('font', family='NanumGothic')
plt.rc('axes', unicode_minus=False)


fig, axes = plt.subplots(2, 2, figsize=(14,10))
fig.suptitle("영화 데이터 분석 시각화", fontsize=16, fontweight='bold')

# (1) 장르별 평점 평균 (막대그래프)
axes[0,0].bar(genre_rating["장르"], genre_rating["평점"], color='royalblue')
axes[0,0].set_title("장르별 평균 평점", fontsize=13)
axes[0,0].set_xlabel("장르")
axes[0,0].set_ylabel("평점")
axes[0,0].grid(alpha=0.3)

# (2) 장르별 평균 관객수 (막대그래프)
axes[0,1].bar(genre_audience["장르"], genre_audience["관객수"], color='lightseagreen')
axes[0,1].set_title("장르별 평균 관객수", fontsize=13)
axes[0,1].set_xlabel("장르")
axes[0,1].set_ylabel("평균 관객수")
axes[0,1].grid(alpha=0.3)

# (3) 영화별 관객수 (선그래프)
axes[1,0].plot(movie_audience["제목"], movie_audience["관객수"],
               marker='o', color='orange', label="관객수")
axes[1,0].set_title("영화별 관객수", fontsize=13)
axes[1,0].set_xlabel("영화제목")
axes[1,0].set_ylabel("관객수")
axes[1,0].tick_params(axis='x', rotation=60)
axes[1,0].legend()
axes[1,0].grid(alpha=0.3)

# (4) 영화별 평점 (선그래프)
axes[1,1].plot(movie_rating["제목"], movie_rating["평점"],
               marker='o', color='purple', label="평점")
axes[1,1].set_title("영화별 평점", fontsize=13)
axes[1,1].set_xlabel("영화제목")
axes[1,1].set_ylabel("평점")
axes[1,1].tick_params(axis='x', rotation=60)
axes[1,1].legend()
axes[1,1].grid(alpha=0.3)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
